Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()



Ans-In Flask, app routing refers to the process of mapping URLs to specific functions or view methods in your application. It involves defining routes using the @app.route() decorator to determine how different URLs should be handled.

The @app.route() decorator is used to associate a URL pattern with a function or view method that will be executed when that URL is requested. It allows you to define different routes for various parts of your application.
By using app routes, you can define the behavior of your Flask application for different URLs, enabling users to access specific content or perform different actions based on the route they visit. This helps in creating a well-structured and interactive web application with multiple pages or endpoints.
